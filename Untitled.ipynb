```python
sns.countplot(x='Gender', data = df)
```

```python
#This code would give the excat count on top of the bar graph

ax = sns.countplot(x='Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#seggregating the amount of purchase based on gender

df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender', y= 'Amount', data = sales_gen)
```

```python
#If we want the amount at the top of the bar plot, then

bx = sns.barplot(x='Gender', y= 'Amount', data = sales_gen)

for bars in bx.containers:
    bx.bar_label(bars)
```

```python
#From above graphs, we can see that women purchase more than men
```

```python
#Now, lets find out which age group purchased more products

sales_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_age
```

```python
sns.barplot(x='Age Group', y= 'Amount', data = sales_age)
```

```python
# From above graphs, we see that women aged between 26-35 had more purchasing power
```

```python
#lets work on states to see from which state did we get the maximum revenue

sales_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)


```

```python
sales_state
```

```python
sns.set(rc={'figure.figsize':(15,5)})

#above line just to set the size of the graph
sns.barplot(data = sales_state, x= 'State', y = 'Orders')
```

```python
#Most of the orders are from UP, Maharashtra and  Karnataka
```

```python
#Now, lets find which product caegory sales was more

sales_pc = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(7)
sns.set(rc = {'figure.figsize':(25,5)})

sns.barplot(data = sales_pc, x= 'Product_Category', y = 'Amount')

```

```python
#Hence people spend more on Food, Clothings and Apparel and Electronic gadgets
```

```python
#Now lets find which product was sold the most

sales_pid = df.groupby(['Product_ID'], as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(7)
sns.set(rc = {'figure.figsize':(25,5)})

sns.barplot(data = sales_pid, x= 'Product_ID', y = 'Amount')
```

```python
#Hence Married women aged 26-35 age group who work in IT industry, Aviation sectors purchase more on Food, Clothing and electronics 
```
